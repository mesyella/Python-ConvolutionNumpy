### 2. Convolution

#### Input image and convert to grayscale

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image, ImageOps 
     
img = Image.open("assets/Sydney-Opera-House.jpg") 
img = ImageOps.grayscale(img) 
plt.imshow(img, cmap="gray")
```

### A. Implement Convolution using Numpy Library

#### Kernel/Filter

```python
img = np.array(img)
filters = np.zeros((2,3,3))
filters[0, :, :] = np.array([[[1/9, 1/9, 1/9],
                            [1/9, 1/9, 1/9],
                            [1/9, 1/9, 1/9]]])
filters[1, :, :] = np.array([[[1, 0, -1],
                            [1, 0, -1],
                            [1, 0, -1]]])
```

#### Convolution Function

```python
def conv(img, conv_filter):  
    filter_size = conv_filter.shape[0]  
    result = np.zeros((img.shape[0]- filter_size + 1,img.shape[1]-filter_size))  
    for r in range (img.shape[0]-filter_size):  
        for c in range(img.shape[1]-filter_size):  
            curr_region = img[r:r+filter_size, c:c+filter_size]  
            curr_result = curr_region * conv_filter  
            conv_sum = np.sum(curr_result)  
            result[r, c] = conv_sum  

    return result 
```

### B. Image Result 

#### Kernel 1

```python
conv1 = conv(img, filters[0])
plt.imshow(conv1, cmap="gray")
```

#### Kernel 2

```python
conv2 = conv(conv1, filters[1])
plt.imshow(conv2, cmap="gray")
```
